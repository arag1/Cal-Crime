# Tracking when Robberies happen

```python
lst = []

for x in df['EVENTTM']:
    hour = x.split(':')[0]
    num_hour = int(hour)
    lst.append(num_hour)
```

# Time Interval between 20:00(8:00 pm) and 4:59 am

```python
df['hour'] = lst
evening = df[(df['hour'] >= 20) | (df['hour'] < 5)]
```

# Time Interval between 5:00 am and 12:59 pm

```python
morning = df[(df['hour'] >= 5) & (df['hour'] < 13)]
```

# Time Interval between 1 pm and 7:59 pm

```python
afternoon = df[(df['hour'] >= 13) & (df['hour'] < 20)]
```

```python
morningframe = morning['OFFENSE'].value_counts().reset_index()
```

```python
morningframe = morningframe.rename(columns={'index':'OFFENSE TYPE', 'OFFENSE':'5am-12:59pm Time Interval'})
```

```python
morningframe
```

```python
afternoonframe = afternoon['OFFENSE'].value_counts().reset_index()
```

```python
afternoonframe = afternoonframe.rename(columns={'index':'OFFENSE TYPE', 'OFFENSE':'1pm-7:59pm Time Interval'})
```

```python
afternoonframe
```

```python
eveningframe = evening['OFFENSE'].value_counts().reset_index()
```

```python
eveningframe = eveningframe.rename(columns={'index':'OFFENSE TYPE', 'OFFENSE':'8pm-4:59am Time Interval'})
```
